```python
import openrouteservice
import json
import folium

API_KEY = "5b3ce3597851110001cf6248ee14ace25f284051b16a4c49fbe9ddff"

# Initialize client
client = openrouteservice.Client(key=API_KEY)

# Define coordinates (longitude, latitude) for Washington, DC and Los Angeles, CA
coordinates = [[-77.0369, 38.9072], [-118.2437, 34.0522]]
locations = ["Washington, DC", "Los Angeles, CA"]

# Request directions
route = client.directions(
    coordinates=coordinates,
    profile='driving-car',
    format='geojson'
)

# Save route to a file
with open("route.json", "w") as file:
    json.dump(route, file, indent=4)

# Extract data
distance = route['features'][0]['properties']['segments'][0]['distance'] / 1000  # Convert to km
duration = route['features'][0]['properties']['segments'][0]['duration'] / 3600  # Convert to hours
best_route = route['features'][0]['geometry']['coordinates']

# Convert route coordinates from [longitude, latitude] to [latitude, longitude] for Folium
route_coords = [(coord[1], coord[0]) for coord in best_route]

# Create map centered at midpoint of route
midpoint = [(coordinates[0][1] + coordinates[1][1]) / 2, (coordinates[0][0] + coordinates[1][0]) / 2]
route_map = folium.Map(location=midpoint, zoom_start=5)

# Add markers for start and end locations
folium.Marker(location=[coordinates[0][1], coordinates[0][0]], popup=locations[0], icon=folium.Icon(color='green')).add_to(route_map)
folium.Marker(location=[coordinates[1][1], coordinates[1][0]], popup=locations[1], icon=folium.Icon(color='red')).add_to(route_map)

# Add route to map
folium.PolyLine(route_coords, color="blue", weight=5, opacity=0.7).add_to(route_map)

# Save map to HTML file and display
route_map.save("route_map.html")

# Print summary
print(f"Route from {locations[0]} ({coordinates[0][1]}, {coordinates[0][0]}) to {locations[1]} ({coordinates[1][1]}, {coordinates[1][0]})")
print(f"Route distance: {distance:.2f} km")
print(f"Estimated duration: {duration:.2f} hours")
print("Best route coordinates:")
for coord in best_route[:5]:  # Display first 5 coordinates as sample
    print(coord)

# Display map in Jupyter/Colab
display(route_map)
```
